```python
import matplotlib.pyplot as plt
import numpy as np
```

# Datos

```python
tiempos=[0,5,7,14,23]
Blanco=[[0.83,0.84],[0.89,0.90,0.85],[0.91,0.87,0.87,0.89],[0.93,0.83],[]]
LL01=[[0.84,0.86],[0.77,0.76,0.77,0.76,0.77,0.78],[0.73,0.76,0.77,0.77],[0.48,0.45,0.46,0.45],[0.26,0.27]]
LL02=[[0.86,0.85],[0.68,0.69,0.68,0.69,0.69,0.7,0.66,0.65],[0.59,0.6,0.63,0.58],[0.08,0.09,0.05,0.03],[0.03,0.02]]
datos=[Blanco,LL01,LL02]
labels=["Blanco","LL01","LL02"]
```

# Calibracion

```python
for d in range(len(datos)):
    for n in range(len(datos[d])):
        for j in range(len(datos[d][n])):
            datos[d][n][j]=(datos[d][n][j]*20/0.18)
```

```python
0.84*0.2
```

# Resultados

```python
# --- Graficar ---
plt.figure(figsize=(8,5))

for serie, etiqueta in zip(datos, labels):
    # Promedios y desvos estndar
    medios = np.array([np.mean(m) for m in serie])
    desvios = np.array([np.std(m, ddof=1) for m in serie])  # ddof=1  std muestral
    
    # Graficar con barras de error
    plt.errorbar(
        tiempos, medios, yerr=desvios,
        fmt='o-', capsize=5, label=etiqueta
    )

plt.xlabel("Tiempo (das)",fontsize=14)
plt.ylabel("Concentracin de MB (g/L)",fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
